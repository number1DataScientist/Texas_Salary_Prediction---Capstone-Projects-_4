# Exploring Relationships:

## Relationship between Gender and Annual Salary:

```python
# Boxplot to show the relationship between gender and annual salary
plt.figure(figsize=(8, 6))
sns.boxplot(x='GENDER', y='ANNUAL', data=data1)
plt.title('Relationship between Gender and Annual Salary')
plt.xlabel('Gender')
plt.ylabel('Annual Salary')
plt.show()
```

```python
# Scatter plot to show the relationship between gender and annual salary
plt.figure(figsize=(8, 6))
sns.scatterplot(x='GENDER', y='ANNUAL', data=data1)
plt.title('Relationship between Gender and Annual Salary')
plt.xlabel('Gender')
plt.ylabel('Annual Salary')
plt.show()
```

# In both the scatter plot and box plot:

- The scatter plot may show a denser concentration of points for females, indicating more female employees compared to males.
- The box plot may have a larger box and potentially wider spread for females, suggesting more variability in the data among female employees compared to males.

## Relationship between Ethnicity and Hourly Rate:

```python
# Boxplot to show the relationship between ethnicity and hourly rate
plt.figure(figsize=(10, 6))
sns.boxplot(x='ETHNICITY', y='HRLY RATE', data=data1)
plt.title('Relationship between Ethnicity and Hourly Rate')
plt.xlabel('Ethnicity')
plt.ylabel('Hourly Rate')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

```python
# Violin plot to show the relationship between ethnicity and hourly rate
plt.figure(figsize=(10, 6))
sns.violinplot(x='ETHNICITY', y='HRLY RATE', data=data1)
plt.title('Relationship between Ethnicity and Hourly Rate')
plt.xlabel('Ethnicity')
plt.ylabel('Hourly Rate')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

# The violin plot for ethnicity and hourly rate shows:

 - 'AM INDIAN' ethnicity has the highest maximum hourly rate among all ethnic groups.
 - Next are 'Asian' individuals with slightly lower hourly rates than 'AM INDIAN'.
 - 'WHITE', 'Hispanic', and 'Black' ethnicities follow in decreasing order of maximum hourly rates.
 - The 'Other' category has a wide range of hourly rates, indicating diversity within this group.
 
This plot helps visualize the variation in hourly rates across different ethnicities, with 'AM INDIAN' showing the highest rates followed by 'Asian', 'WHITE', 'Hispanic', 'Black', and 'Other' groups. 

## Relationship between Status and Annual Salary:

```python
# Box plot to show the relationship between status and annual salary
plt.figure(figsize=(10, 6))
sns.boxplot(x='ANNUAL', y='STATUS', data=data1)
plt.title('Relationship between Status and Annual Salary')
plt.xlabel('Annual Salary')
plt.ylabel('Status')
plt.show()
```

```python
# scatterplot to show the distribution of annual salaries across different status categories
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ANNUAL', y='STATUS', data=data1)
plt.title('Distribution of Annual Salaries by Status')
plt.xlabel('Annual Salary')
plt.ylabel('Status')
plt.show()
```

# In both the scatter plot and box plot:

 - The scatter plot may exhibit a trend where higher-status categories (such as ERF and URF) are positioned at the top, indicating that individuals in these statuses tend to receive higher annual salaries compared to individuals in lower-status categories.
 - Similarly, the box plot may illustrate that the median salary is higher for statuses such as ERF and URF, gradually decreasing for lower-status categories, thus reflecting the hierarchy of annual salaries across different statuses.

## Employment Trends Over Time

```python
# Convert 'EMPLOY DATE' to datetime if needed
data1['EMPLOY DATE'] = pd.to_datetime(data1['EMPLOY DATE'])

# Extract year from 'EMPLOY DATE' and analyze employment trends over time
data1['EMPLOY YEAR'] = data1['EMPLOY DATE'].dt.year
employment_trends = data1['EMPLOY YEAR'].value_counts().sort_index()

# Plot employment trends over time
plt.figure(figsize=(10, 6))
plt.plot(employment_trends.index, employment_trends.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Employees')
plt.title('Employment Trends Over Time')
plt.grid(True)
plt.show()
```

# Employment Trends Over Time
 - The "Employment Trends Over Time" analysis provides a visual representation of how the number of employees has changed over different years. It helps identify trends, spikes, or dips in employment levels, offering insights into hiring patterns and workforce changes over time within the dataset.

# Count Plots for Categorical Variables

```python
categorical_columnss = ['ETHNICITY', 'GENDER', 'STATUS']

# Create subplots for count plots of each categorical column
fig, axes = plt.subplots(nrows=len(categorical_columnss), ncols=1, figsize=(10, 6 * len(categorical_columnss)))
for i, column in enumerate(categorical_columnss):
    sns.countplot(y=column, data=data1, ax=axes[i])
    axes[i].set_title(f'Count Plot of {column}')
    axes[i].set_xlabel('Count')
    axes[i].set_ylabel(column)
    axes[i].set_xlim(0, len(data1))  # Set x-axis limit based on the data size

plt.tight_layout()
plt.show()
```
